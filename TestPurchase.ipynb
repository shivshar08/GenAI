```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
y_train.shape
```

```python
# Scale a Feature to kill those varying values between different scales of features
# Age and Estimated Salary
from sklearn.preprocessing import StandardScaler
#After the transformation - The mean of the feature/column=0, std=-1

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

```

```python
X_train
```

```python
# Binary Classification: Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
# Predict Purchase
# 30, 300000
print(classifier.predict(sc.transform([[30,300000]])))
```
